```python
list1 = [12, 23, 34, 45, 56, 67, 78, 89, 100]
list2 = [21, 32, 43, 54, 65, 76, 87, 98, 100]


plt.xlabel('List - 1')
plt.ylabel('List - 2')

plt.title('Scatter Chart')

plt.scatter(list1, list2)
```

```python
prdt_id = [1, 2, 3, 4, 5, 6, 7, 8]
prdt_online_price = [201, 114, 125, 115, 198, 146, 77, 120]
prdt_offline_price = [155, 100, 120, 90, 102, 140, 180, 102]
prdt_thiefMarket_price = [250, 150, 122, 50, 150, 145, 151, 111]
```

```python
plt.plot(prdt_id, prdt_online_price, color = 'green', marker = '*', markersize = 5)

plt.xlabel('Product ID')
plt.ylabel('Product Online Price')


plt.title('Product vs Online Price')
```

```python
plt.plot(prdt_id, prdt_online_price, marker = '*', markersize = 5, label = 'Online Price')
plt.plot(prdt_id, prdt_offline_price, color = 'green', marker = 'D', markersize = 5, label = 'Offline Price')
plt.plot(prdt_id, prdt_thiefMarket_price, color = 'red', marker = '+', markersize = 5, label = 'Thief Market Price')

plt.xlabel('Product ID')
plt.ylabel('Product Online Price')

plt.legend()

plt.title('Product vs Price')
```
