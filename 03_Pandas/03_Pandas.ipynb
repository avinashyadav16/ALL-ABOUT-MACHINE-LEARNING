## The isin() Function


- Find out the most destructive death overs(last five overs) batsman in the history if IPL.
- The batsman has played Minimum of 200 Balls.

- Strike Rate = (Number of Runs / Number of Balls) \* 100


```python
# The data about the death overs:
death_overs = delivery_df[delivery_df['over'] > 15]
```

```python
# Number of balls played by the players:
delivery_df.groupby('batter')['batsman_runs'].count()
```

```python
# Now counting the total number of deliveries played by the batsman in the death overs:
death_over_batters = death_overs.groupby('batter')['batsman_runs'].count()
death_over_batters
```

```python
# Now getting only those batsman who has played more than 200 balls
more_than_200_balls = death_over_batters > 200
```

```python
# Doing Boolean Indexing to get the names of the batters:
death_over_batters[more_than_200_balls]
```

```python
death_over_batters[more_than_200_balls].index
```

```python
batsman_list = death_over_batters[more_than_200_balls].index.tolist()
```

```python
# Now we need to calculate the strike rate of all the batters, for that we need 2 things:
# 1. Runs Scored By Them and
# 2. Number of Balls Played By Them

delivery_df['batter'].isin(batsman_list)
```

```python
final = delivery_df[delivery_df['batter'].isin(batsman_list)]
```

```python
runs = final.groupby('batter')['batsman_runs'].sum()
```

```python
balls = final.groupby('batter')['batsman_runs'].count()
```

```python
sr = (runs / balls) * 100
sr.sort_values(ascending=False).head(10)
```

## The Merge() Function:


- List of Orange Cap Holder Batter:

- Both the data sets have same values for the columns 'id' from match_df and 'match_id' from the delivery_df


```python
match_df.head(1)
```

```python
delivery_df.head(1)
```

```python
print(match_df.shape)
print(delivery_df.shape)
```

```python
merged_df = delivery_df.merge(match_df, left_on='match_id', right_on='id')
merged_df.head(1)
```

```python
# First grouping based on the season the followed by the batter:
merged_df.groupby(['season', 'batter'])['batsman_runs'].sum()
```

```python
# Converting the series into dataframes and getting the top scorers:
merged_df.groupby(['season', 'batter'])['batsman_runs'].sum(
).sort_values(ascending=False).reset_index()
```

```python
# Dropping the subsequent top scorer in any season if they are present:
merged_df.groupby(['season', 'batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')
```

```python
# Orange Cap Holders are:
merged_df.groupby(['season', 'batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')[['season', 'batter']]
```
